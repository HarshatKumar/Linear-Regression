```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load data
df = pd.read_csv("1000_Companies.csv")

# Visualize R&D Spend and Administration using bar plots
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
sns.barplot(x='State', y='R&D Spend', data=df)
plt.title('R&D Spend by State')

plt.subplot(2, 1, 2)
sns.barplot(x='State', y='Administration', data=df)
plt.title('Administration Spend by State')

plt.tight_layout()
plt.show()

# Perform linear regression
X = df[['R&D Spend']]
y = df['Profit']

regression = LinearRegression()
regression.fit(X, y)

# Predict for a sample value
predicted_profit = regression.predict([[165349.20]])
print("Predicted Profit:", predicted_profit)

# Add two more visualizations for predicted data
# For example, scatter plot and residual plot

# Scatter plot of actual vs. predicted profit
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y, y=regression.predict(X))
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs. Predicted Profit')
plt.show()

# Residual plot
residuals = y - regression.predict(X)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=regression.predict(X), y=residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted Profit')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
